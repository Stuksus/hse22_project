```python
ZH_EXECUTABLE = Path("/content/zhunt3")
assert ZH_EXECUTABLE.is_file()


def zhunt(query: str, windowsize: int = 6, minsize: int = 3, maxsize: int = 6):
    query = query.upper()
    assert set(query).issubset({"A", "C", "G", "T", "N"})
    fd, temp = tempfile.mkstemp()
    os.close(fd)
    with open(temp, 'w') as stream:
        stream.write(query)

    subprocess.run(
        [ZH_EXECUTABLE, 
         str(windowsize), str(minsize), str(maxsize), temp],
        check=True, stdout=PIPE, stderr=DEVNULL,
        input=query, encoding='ascii'
    )
    with open(temp + ".Z-SCORE", 'r') as stream:
        df = pd.read_csv(stream,
                         names=['Start', 'End', 'nu-1', 'nu-2', 'nu-3', 
                                'ZH-Score', 'Sequence', 'Conformation'],
                         skiprows=1, sep='\s+')
    os.remove(temp)
    os.remove(temp + ".Z-SCORE")
    return df[['Start', 'End', 'ZH-Score', 'Sequence', 'Conformation']]
```

```python
from Bio import SeqIO

fna_files = ["/content/GCF_000209065.1_ASM20906v1_genomic.fna",
             "/content/GCA_022059095.1_T.mel.1_genomic.fna",
             "/content/GCA_002087225.1_Tth_v3_genomic.fna",
             "/content/GCA_021307395.1_TvIL_AH_1392_genomic.fna",
             "/content/GCA_003719475.1_ASM371947v1_genomic.fna"]

df_list = list()
name_list = list()
for f in fna_files:
    with open(f, 'r') as stream:
        sequences = list(SeqIO.parse(stream, format='fasta'))
    for s in sequences:
        print(f"{s.description} \n\t=> {len(s)}")
    
    seq = sequences[0]
    name_list += [seq.name]
    df = zhunt(str(seq.seq))

    df_list.append(df)
```

```python
def make_bed(data,fileName, genome_file ):
  
  to_write = data.copy()
  to_write.insert(0, 'chr', str(genome_file).split('.')[0])
  to_write['name'] =to_write.chr.astype(str) +'_'+ to_write.Start.astype(str)
  to_write[['chr','Start','End','name','ZH-Score']].to_csv(fileName, sep = '\t', header=False,index=False)
```

```python
d = frame[(frame['# Species'] == 5) & (frame.Genes == 5)].sample(200,random_state = 42)
feature_tables = ['GCA_002087225.1_Tth_v3_feature_table.txt',
                  'GCA_003719475.1_ASM371947v1_feature_table.txt',
                  'GCA_021307395.1_TvIL_AH_1392_feature_table.txt',
                  'GCA_022059095.1_T.mel.1_feature_table.txt',
                  'GCF_000209065.1_ASM20906v1_feature_table.txt']
best_clust = []
for i in tqdm(range(0, d.shape[0])):
    row = d.iloc[i]
    genes = [row['protein.GCA_002087225.faa'], row['protein.GCA_003719475.faa'], row['protein.GCA_021307395.faa'], row['protein.GCA_022059095.faa'], row['protein.GCF_000209065.faa']]
    sum_intersec = 0
    for j in range(5):
        f = pd.DataFrame(columns=['name', 'start', 'end', 'n'])
        table = pd.read_csv(feature_tables[j], sep='\t')
        row_gene = table.loc[table[table['product_accession'] == genes[j]].index - 1]
        if row_gene['strand'].values[0] == '+':
            pos = row_gene['start'].values[0]
        else:
            pos = row_gene['end'].values[0]
        f.loc[0] = [names_fna[j], pos, pos, names_fna[j] + '_' + str(pos)]
        f.to_csv('temp.bed', sep = '\t', header=False, index=False)
        len_name = names_wo[j] + '_len.bed'
        ! bedtools slop -i temp.bed -g '{len_name}' -b 300 > temp_slop.bed
        ! intersectBed -a '{names_final_bed[j]}' -b temp_slop.bed -wb > temp_intersec.bed
        intersec_number = ! wc -l temp_intersec.bed
        sum_intersec += int(intersec_number[0].split(' ')[0])

    best_clust.append((i, sum_intersec))


```

```python
best_clusters = sorted(best_clust, key=lambda x: -x[1])[0:10]
cluster_ind = list(map(lambda x: x[0], best_clusters))
clusters = d.iloc[cluster_ind]
```

```python
scores = []
scores_for_genome = []
for i in tqdm(range(0, clusters.shape[0])):
    row = clusters.iloc[i]
    genes = [row['protein.GCA_002087225.faa'], row['protein.GCA_003719475.faa'], row['protein.GCA_021307395.faa'], row['protein.GCA_022059095.faa'], row['protein.GCF_000209065.faa']]
    sum_intersec = 0
    genome_score = []
    for j in range(5):
        f = pd.DataFrame(columns=['name', 'start', 'end', 'n'])
        table = pd.read_csv(feature_tables[j], sep='\t')
        row_gene = table.loc[table[table['product_accession'] == genes[j]].index - 1]
        if row_gene['strand'].values[0] == '+':
            pos = row_gene['start'].values[0]
        else:
            pos = row_gene['end'].values[0]
        f.loc[0] = [names_fna[j], pos, pos, names_fna[j] + '_' + str(pos)]
        f.to_csv('temp.bed', sep = '\t', header=False, index=False)
        len_name = names_wo[j] + '_len.bed'
        ! bedtools slop -i temp.bed -g '{len_name}' -b 300 > temp_slop.bed
        ! intersectBed -a '{names_final_bed[j]}' -b temp_slop.bed -wb > temp_intersec.bed
        lines = ! wc -l temp_intersec.bed
        if int(lines[0].split(' ')[0]) == 0:
            continue
        t = pd.read_csv('temp_intersec.bed', sep='\t', header = None)
        genome_score.append(t.loc[:, 4].sum())
    scores_for_genome.append(np.round(genome_score, 2))
    scores.append(np.round(np.mean(genome_score), 2))
```

```python
clusters_final = clusters.copy()
clusters_final.loc[:, 'zh-score'] = scores_for_genome
clusters_final.loc[:, 'mean_zh-score'] = scores
clusters_final = clusters_final.reset_index()
```

```python
for i in tqdm(range(clusters_final.shape[0])):
    to_write = []
    row = clusters_final.loc[i]
    for j in range(5):
        protein = row[clusters_final.columns[j + 4]]
        file_name = names_wo[j] + '_protein'

        faa_file = open(clusters_final.columns[j + 4], "r")
        flag = False

        while True:
            line = faa_file.readline()
            if not line:
                break
            if '>' in line.strip():
                flag = False
            if protein in line.strip():
                flag = True
            if flag:
                to_write.append(line.strip())
        faa_file.close()
    f = open('cluster' + str(i) + '.txt', 'w' )
    f.write('\n'.join(to_write))
    f.close()
```
